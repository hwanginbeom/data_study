```python
import pandas as pd
from haversine import haversine
```

# 1. 데이터 불러오기

```python
# hangjeongdong + 중심 위도 &경도
cent_hangjeongdong = pd.read_csv('C:/Users/user/Desktop/도서나루/행정동 데이터.csv', index_col=[0])
cent_hangjeongdong.head()
```

```python
# seoul_libry
seoul = pd.read_csv('C:/Users/user/Desktop/도서나루/seoul_libry.csv', index_col=[0])
seoul.head()
```

# 2. 행정동 별 공공 도서관 없는 지역 찾기

```python
seoul1 = seoul[['행정동','lib_name']]
seoul1 = seoul1.groupby(seoul1['행정동']).count().reset_index()
seoul1.columns = ['행정동', 'lib_cnt']
seoul1.head()
```

```python
pd.merge(seoul1, cent_hangjeongdong, on='행정동', how='right')
```

```python
dong_library = pd.merge(seoul1, cent_hangjeongdong, on='행정동', how='right').fillna(0)
dong_not_library = dong_library[dong_library['lib_cnt']==0]
dong_not_library.shape
```

```python
dong_not_library
```

# 3. 공공도서관과의 최소 거리

- 공공도서관이 없는 행정동의 중심좌표에서 가장 가까운 도서관과의 최소 거리 구하기

```python
dong_not_library = dong_not_library.reset_index(drop=True)
dong_not_library.head()
```

```python
dong_not_library.shape
```

```python
seoul_points = seoul[['lib_latitude', 'lib_longitude']].values
seoul_points
```

```python
dong_not_library['최소거리'] = 0

for i in range(len(dong_not_library)):
    target = target = dong_not_library.iloc[i][['cent_lat', 'cent_lon']].values
    
    answer = 100000
    for point in seoul_points:
        result = haversine(target, point, unit='m')
        answer = min(answer, result)
        print(answer)
        
    dong_not_library.loc[i, '최소거리'] = answer
```

```python
# 본인 경로
dong_not_library.to_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/최소거리필터링.csv', encoding='utf-8-sig', index=False)
```
