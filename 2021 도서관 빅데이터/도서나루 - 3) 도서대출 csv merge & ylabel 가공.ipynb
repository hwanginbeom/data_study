# 1. 크롤링 데이터

### 1) 데이터 불러오기

```python
page_1 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/1page.csv', encoding='utf-8')
page_2 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/2page.csv', encoding='utf-8')
page_3 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/3page.csv', encoding='cp949', sep=',')
page_4 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/4page.csv', encoding='cp949', sep=',')
page_5 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/5page.csv', encoding='utf-8')
page_6 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/6page.csv', encoding='cp949', sep='\t')
page_7 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/7page.csv', encoding='utf-8')
page_8 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/8page.csv', encoding='utf-8')
page_9 = pd.read_csv('C:/Users/user/Desktop/도서나루/도서관별 대출도서/9page.csv', encoding='utf-8')
```

```python
page_1.columns = ['xlsx_name', 'total_num']
```

### 2) 데이터 결합

```python
data = pd.concat([page_1, page_2, page_3, page_4, page_5, page_6, page_7, page_8, page_9], axis=0)
data = data.reset_index(drop=True)
data.shape
```

```python
data
```

### 3) 데이터 가공

```python
for i in range(len(data)):
    if data.loc[i,'xlsx_name'].find('평생학습관') != -1:
        data.loc[i, 'library_name'] = data.loc[i, 'xlsx_name'][:data.loc[i,'xlsx_name'].find('평생학습관')] + '평생학습관'
    else:
        data.loc[i, 'library_name'] = data.loc[i, 'xlsx_name'][:data.loc[i,'xlsx_name'].find('도서관')] + '도서관'
```

```python
# data['library_name'].unique()
```

```python
for i in range(len(data)):
    date = data.loc[i, 'xlsx_name'][data.loc[i,'xlsx_name'].find('(')+1 : data.loc[i,'xlsx_name'].find(')')]
    data.loc[i, 'date'] = date.replace(' ','').replace('년','').replace('월','')
```

```python
result.isnull().sum()
```

- 찾지 못한 나머지 결측 데이터는 삭제

```python
result = result.dropna()
del result['library_name']
result.shape
```

# 3. y_label 가공

### 1) y_label(총 대출건수)가 누적 데이터인지 확인

```python
data1 = pd.read_excel('C:/Users/user/Desktop/도서나루/도서관별 대출도서/7page/송파글마루도서관 장서 대출목록 (2019년 01월).xlsx')
data2 = pd.read_excel('C:/Users/user/Desktop/도서나루/도서관별 대출도서/7page/송파글마루도서관 장서 대출목록 (2019년 02월).xlsx')
```

```python
# 하나의 도서가 여러번 적힌 경우 발생
print('data1 - 전체 데이터 형태 :', data1.shape)
print('data1 - 데이터 내 도서명 개수 :', len(data1['도서명'].unique()))
print()
print('data2 - 전체 데이터 형태 :', data2.shape)
print('data2 - 데이터 내 도서명 개수 :', len(data2['도서명'].unique()))
```

```python
# 도서명으로 대출건수 다시 그루핑해서 계산
d1 = pd.DataFrame(data1['대출건수'].groupby(data1['도서명']).sum())
d2 = pd.DataFrame(data2['대출건수'].groupby(data2['도서명']).sum())
```

```python
df = pd.merge(d1, d2, how='outer', on='도서명')
```

```python
df.isnull().sum()
```

```python
# y_label이 누적 데이터 임을 확인
df = df.fillna(0)
df[df['대출건수_x']>df['대출건수_y']]
```
