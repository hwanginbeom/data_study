```python
from tqdm import tqdm_notebook
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import requests
import urllib
import time

from random import *
import numpy as np
import pandas as pd
```

## 크롤링

```python
driver = webdriver.Chrome("C:/chromedriver.exe")
url = "https://www.data4library.kr/openDataL"
driver.get(url)
```

```python
# 서울/공공 선택해서 검색
select1 = Select(driver.find_element_by_id("p_region"))
select1.select_by_visible_text('서울')
select2 = Select(driver.find_element_by_id("libType"))
select2.select_by_visible_text('공공')
driver.find_element_by_class_name('ls_search_bt').click()
```

```python
driver.execute_script('goPage(2)')
elements = driver.find_elements_by_class_name('link_td')
```

```python
for i in range(len(elements)):
    print(i)
    
    url = "https://www.data4library.kr/openDataL"
    driver.get(url)
    
    # 서울/공공 선택해서 검색
    select1 = Select(driver.find_element_by_id("p_region"))
    select1.select_by_visible_text('서울')
    select2 = Select(driver.find_element_by_id("libType"))
    select2.select_by_visible_text('공공')
    driver.find_element_by_class_name('ls_search_bt').click()
    driver.execute_script('goPage(2)')
    
    elements = driver.find_elements_by_class_name('link_td')    
    elements[i].click()
    time.sleep(3)
    
    page_bar = driver.find_element_by_css_selector('#sb-site > section > div.sub_container > div.paging_nav.mgb_20')
    page = page_bar.find_elements_by_css_selector('a')
    last_page = page[-3].text
    if page[-3].text.isdigit() == False:
        table = driver.find_element_by_css_selector('#sb-site > section > div.sub_container > div:nth-child(5) > table > tbody')
        excel_list = table.find_elements_by_class_name('download_link.excel_type')
        for j in range(len(excel_list)):
            time.sleep(3)
            print(j)
            excel_list[j].click()
        continue
            
    for i in range(1, int(last_page)+1):
        print('goPage(' + str(i) + ')')
        driver.execute_script('goPage(' + str(i) + ')')
    
        table = driver.find_element_by_css_selector('#sb-site > section > div.sub_container > div:nth-child(5) > table > tbody')
        excel_list = table.find_elements_by_class_name('download_link.excel_type')
        for j in range(len(excel_list)):
            time.sleep(4)
            print(j)
            excel_list[j].click()
```
